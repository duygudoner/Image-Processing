```python
im_2=np.zeros((256,256),dtype=np.uint8)
```

```python
im_2.shape
```

```python
im_2=im_1[:,:,0] # siyah-beyaz resim
im_3=im_1[:,:,0] # filtreleme yapılacak resim
im_3=np.zeros((256,256),dtype=np.uint8)
```

```python
plt.imshow(im_2,cmap='gray')
plt.show()
```

```python
m,n=im_2.shape
```

```python
for i in range(1,m-1):
    for j in range(1,n-1):
        s= \
        im_2[i-1,j-1]/9+ \
        im_2[i-1,j]/9+ \
        im_2[i-1,j+1]/9+ \
        im_2[i,j-1]/9+ \
        im_2[i,j]/9+ \
        im_2[i,j+1]/9+ \
        im_2[i+1,j-1]/9+ \
        im_2[i+1,j]/9+ \
        im_2[i+1,j+1]/9 
        s=int(s)
        # print(s ,end=' * ')
        im_3[i,j]=s
```

```python
plt.subplot(1,2,1)  # resmi konumlandırma yanyana (ilk sütun)
plt.imshow(im_2,cmap='gray')
plt.subplot(1,2,2) # (ikinci sütun)
plt.imshow(im_3,cmap='gray')
```
